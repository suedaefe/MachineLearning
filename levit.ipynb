```python
import os
import torch
import torchvision.transforms as T
import torchvision.datasets as datasets
from torch.utils.data import DataLoader, SubsetRandomSampler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve, auc, roc_auc_score, confusion_matrix, matthews_corrcoef
from transformers import LevitForImageClassification, LevitFeatureExtractor
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torch.cuda.amp import GradScaler, autocast

DATASET_PATH = r'C:\Users\sueda\OneDrive\Masaüstü\makina'
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f'Using device: {device}')

feature_extractor = LevitFeatureExtractor.from_pretrained('facebook/levit-384')

transform_train = T.Compose([
    T.Resize((384, 384)),
    T.RandomHorizontalFlip(),
    T.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4, hue=0.3),
    T.RandomRotation(degrees=30),
    T.RandomResizedCrop(size=(384, 384), scale=(0.6, 1.0)),
    T.ToTensor(),
    T.Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)
])

transform_test = T.Compose([
    T.Resize((384, 384)),
    T.ToTensor(),
    T.Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)
])

dataset_train = datasets.ImageFolder(DATASET_PATH, transform=transform_train)
dataset_test = datasets.ImageFolder(DATASET_PATH, transform=transform_test)
class_names = dataset_train.classes
num_classes = len(class_names)
print(f'Classes: {class_names}')

def initialize_model(num_labels):
    model = LevitForImageClassification.from_pretrained(
        'facebook/levit-384',
        num_labels=num_labels,
        ignore_mismatched_sizes=True
    )

    model.classifier.dropout = nn.Dropout(p=0.6)

    return model.to(device)

def train(model, dataloader, criterion, optimizer, scaler, gradient_accumulation_steps=4):
    model.train()
    total_loss = 0
    correct = 0
    total = 0
    
    optimizer.zero_grad() 
    
    for step, (inputs, labels) in enumerate(dataloader):
        inputs, labels = inputs.to(device), labels.to(device)
        
        with autocast():
            outputs = model(inputs).logits 
            loss = criterion(outputs, labels)
            total_loss += loss.item()
        
        scaler.scale(loss).backward()  
        
        if (step + 1) % gradient_accumulation_steps == 0 or (step + 1) == len(dataloader):
            scaler.step(optimizer) 
            scaler.update()
            optimizer.zero_grad()  

        _, predicted = torch.max(outputs, 1)  
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total
    return avg_loss, accuracy

def evaluate(model, dataloader, criterion):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    
    all_labels = []
    all_probs = []
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs).logits 
            loss = criterion(outputs, labels)
            
            total_loss += loss.item()
            _, predicted = torch.max(outputs, 1) 
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            
            all_probs.extend(torch.softmax(outputs, dim=1).cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(dataloader)
    accuracy = correct / total
    return avg_loss, accuracy, np.array(all_labels), np.array(all_probs)


def calculate_metrics(y_true, y_pred, y_probs, num_classes, class_names):
    metrics = {}
    for i in range(num_classes):
        class_metrics = {}
        tp = ((y_true == i) & (y_pred == i)).sum()
        tn = ((y_true != i) & (y_pred != i)).sum()
        fp = ((y_true != i) & (y_pred == i)).sum()
        fn = ((y_true == i) & (y_pred != i)).sum()
        
        precision = tp / (tp + fp) if (tp + fp) > 0 else 0
        recall = tp / (tp + fn) if (tp + fn) > 0 else 0
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        accuracy = (tp + tn) / (tp + tn + fp + fn)
        mcc = matthews_corrcoef(y_true == i, y_pred == i)
        roc_auc = roc_auc_score(y_true == i, y_probs[:, i])
        
        class_metrics['Precision'] = precision
        class_metrics['Recall'] = recall
        class_metrics['F1-score'] = f1_score
        class_metrics['Accuracy'] = accuracy
        class_metrics['Sensitivity (Recall)'] = recall
        class_metrics['Specificity'] = specificity
        class_metrics['MCC'] = mcc
        class_metrics['AUC'] = roc_auc
        
        metrics[class_names[i]] = class_metrics
    
    return metrics

def k_fold_cross_validation(k, dataset_train, dataset_test, num_classes, class_names, batch_size, epochs, num_workers):
    kf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
    fold_results = []

    all_indices = np.arange(len(dataset_train))
    all_labels = np.array([dataset_train.targets[i] for i in all_indices])

    overall_train_losses = np.zeros(epochs)
    overall_test_losses = np.zeros(epochs)
    overall_train_accuracies = np.zeros(epochs)
    overall_test_accuracies = np.zeros(epochs)

    all_fprs = {i: [] for i in range(num_classes)}
    all_tprs = {i: [] for i in range(num_classes)}
    mean_fpr = np.linspace(0, 1, 100)

    for fold, (train_idx, test_idx) in enumerate(kf.split(all_indices, all_labels)):
        print(f'Fold {fold + 1}/{k}')
        
        train_sampler = SubsetRandomSampler(train_idx)
        test_sampler = SubsetRandomSampler(test_idx)
        
        train_loader = DataLoader(dataset_train, batch_size=batch_size, sampler=train_sampler, num_workers=num_workers)
        test_loader = DataLoader(dataset_test, batch_size=batch_size, sampler=test_sampler, num_workers=num_workers)
        
        model = initialize_model(num_classes)
        criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
        optimizer = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=0.01)
        scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.5)
        scaler = GradScaler()
        
        train_losses, test_losses = [], []
        train_accuracies, test_accuracies = [], []
        
        for epoch in range(epochs):
            train_loss, train_accuracy = train(model, train_loader, criterion, optimizer, scaler)
            test_loss, test_accuracy, _, _ = evaluate(model, test_loader, criterion)
            
            train_losses.append(train_loss)
            test_losses.append(test_loss)
            train_accuracies.append(train_accuracy)
            test_accuracies.append(test_accuracy)
            
            print(f'Epoch {epoch + 1}/{epochs}: '
                  f'Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.4f}, '
                  f'Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}')
            
            scheduler.step()  
        
        overall_train_losses += np.array(train_losses)
        overall_test_losses += np.array(test_losses)
        overall_train_accuracies += np.array(train_accuracies)
        overall_test_accuracies += np.array(test_accuracies)

        test_loss, test_accuracy, y_true, y_probs = evaluate(model, test_loader, criterion)

        for i in range(num_classes):
            fpr, tpr, _ = roc_curve(y_true == i, y_probs[:, i])
            all_fprs[i].append(np.interp(mean_fpr, fpr, tpr))
            all_fprs[i][-1][0] = 0.0
            all_tprs[i].append(tpr)
        
        fold_results.append({
            'train_losses': train_losses,
            'test_losses': test_losses,
            'train_accuracies': train_accuracies,
            'test_accuracies': test_accuracies
        })

        plt.figure(figsize=(8, 6))
        plt.plot(range(1, epochs + 1), train_losses, label='Train Loss')
        plt.plot(range(1, epochs + 1), test_losses, label='Test Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.title(f'Fold {fold + 1}: Loss per Epoch')
        plt.legend()
        plt.show()

        y_pred = np.argmax(y_probs, axis=1)
        metrics = calculate_metrics(y_true, y_pred, y_probs, num_classes, class_names)
        fold_results.append({
            'train_losses': train_losses,
            'test_losses': test_losses,
            'train_accuracies': train_accuracies,
            'test_accuracies': test_accuracies,
            'metrics': metrics
        })
        
        for class_name, class_metrics in metrics.items():
            print(f'Fold {fold + 1} - Class {class_name} Metrics:')
            for metric, value in class_metrics.items():
                print(f'  {metric}: {value:.4f}')
    
    overall_train_losses /= k
    overall_test_losses /= k
    overall_train_accuracies /= k
    overall_test_accuracies /= k

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, epochs + 1), overall_train_losses, label='Overall Train Loss')
    plt.plot(range(1, epochs + 1), overall_test_losses, label='Overall Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Overall Train vs Test Loss per Epoch')
    plt.legend()
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.plot(range(1, epochs + 1), overall_train_accuracies, label='Overall Train Accuracy')
    plt.plot(range(1, epochs + 1), overall_test_accuracies, label='Overall Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Overall Train vs Test Accuracy per Epoch')
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 8))
    for i in range(num_classes):
        mean_tpr = np.mean(all_fprs[i], axis=0)
        mean_tpr[-1] = 1.0
        mean_auc = auc(mean_fpr, mean_tpr)
        plt.plot(mean_fpr, mean_tpr, label=f'Class {class_names[i]} (AUC = {mean_auc:.2f})')
    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Average ROC Curve')
    plt.legend(loc='lower right')
    plt.show()

    return fold_results

results = k_fold_cross_validation(5, dataset_train, dataset_test, num_classes, class_names, batch_size=16, epochs=15, num_workers=4)
```
